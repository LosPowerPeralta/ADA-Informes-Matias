#1. Descripción del problema

En la teoría de la complejidad computacional los problemas $NP$ y $P$ son una clase de complejidad los cuales nos ayudan a calificar los tipos de problemas a los que nos estamos enfrentando.

Los problemas $P$ son los más simples en describir ya que son aquellos que se pueden resolver en un tiempo polinomial, como por ejemplo los problemas de ordenamiento, de busqueda, entre otros. En cambio, los problemas $NP$ se dificultan mucho más a la hora de describirlos ya que estos se dividen en 3 secciones. La primera sección son los poblemas $NP$, estos pertenecen al conjunto de problemas que su solución puede ser verificada en un tiempo polinomial pero el tiempo que toma resolverlo es exponencial, la segunda sección son los problemas $NP-Hard$, este tipo de problemas son muy dificiles de resolver y para casos de prueba pequeños toma mucho tiempo encontrar una solución clara, además sus soluciones no pueden ser verificadas en un tiempo polinomial, y por último en la tercera sección nos encontramos con los problemas $NP-Complete$, estos problemas son una combinación de los problemas $NP$ y $NP-Hard$ ya que sus soluciones pueden ser verificadas en un tiempo polinomial pero si intentamos reducir un problema $NP-Hard$ con un problema $NP$, esto quiere decir que el problema $NP$ es tan difícil de resolver como un $NP-Hard$, por esta misma razón toma la intersección.

###1. Que quiere decir que un problema sea NP-Hard?

Como habíamos dicho previamente, los problemas $NP-Hard$ son aquellos considerados muy difíciles de resolver y además si es que logramos encontrar una solución a un problema propuesto no podemos verificar esa solución de forma polinomial, a no ser que este problema sea a la vez $NP-Complete$ que estos problemas si se pueden comprobar en tiempo polinomial pero son tan difíciles como un problema $NP-Hard$.

###2. Que quiere decir que $P \neq NP$?

Luego de haber definido los problemas $P$ y $NP$, ahora nos toca hablar sobre la diferencia que hay entre estos dos actualmente, pero, por que "actualmente"?, esto ocurre ya que en estos días aún no se ha logrado demostrar que realmente estos problemas sean distintos del todo pero se acepta que realmente es así, conviertiendo esta frase en una conjetura.

Luego de explicar todo esto, $P \neq NP$ quiere decir que hasta el momento se han descubierto problemas los cuales el tiempo que se demoran en encontrar una solución es exponencial pero no se han encontrado formas de disminuir este tiempo a polinomial, en cambio si algun problema del conjunto $NP$ se lograse reducir a un problema $P$, esto querría decir que los problemas $P$ son iguales a los $NP$, por lo tanto rompería con la conjetura y quedaría demostrado que $P = NP$.

###3. Como se comprueba que un problema es NP-Hard usando reducciones? 

Para comprobar que un problema es $NP-Hard$ utilizando reducciones, primero necesitamos de un problema $A \in NP-Hard$ que pueda ser reducido a un problema $B$ que es el que queremos demostrar, ahora, si el problema $B$ permite reducir el problema $A$ esto quiere decir que el problema $B$ es tan difícil como el problema $A$, por lo tanto esto comprueba que el problema $B$ también es $NP-Hard$.

### **Problema de la suma de subconjuntos (SSP)**

Existen distintas variantes del problema, tales como la variante que acepta solo enteros positivos, una que acepta de ambos signos en la que el valor objetivo es 0 u otra en la que el valor objetivo es la mitad de la sumatoria de todos los valores. También podemos clasificar la variante por la salida del algoritmo, si la salida es $SÍ$ o $NO$, la variante es un **problema de decisión**, pero si la salida es todo el subconjunto que sumado es igual al valor objetivo, entonces es un **problema de optimización**.

La variante a continuación es la de decisión que acepta solo enteros positivos para los valores del conjunto y el valor objetivo.

---
**Entrada**: Conjunto de valores enteros positivo $V = \{v_1, v_2,..., v_i\}$ y un valor objetivo entero positivo $T$.

**Salida**:  Un enunciado que declare $SÍ$ en caso de que exista un subconjunto cuya suma sea el valor objetivo, y $NO$ en caso contrario.

---

### **Teorema**
>*El problema de la suma de subconjuntos es **NP-Hard**.*

Como se explicó anteriormente, para demostrar que el problema de la suma de subconjuntos es $NP-Hard$, debemos lograr reducir un problema $A \in NP-Hard$ al problema SSP. El problema $A$ escogido es el problema de decisión de conjuntos independientes (IS) y asumiremos que este es $NP-Hard$.

### **Lema**

### **Prueba del lema**
Antes de empezar se describirá el problema de decisión de conjuntos independientes.

Dado un grafo no dirigido, determinar si existe un subconjunto independiente de tamaño $k$. Se define como independiente al conjunto de nodos de un grafo en los cuales no hay dos nodos adyacentes, es decir, que los conecte algún arco.

---
**Entrada**: Un grafo no dirigido con un conjuntos de nodos $V$ y de arcos $E$, además de un valor $k$, que será el tamaño del conjunto independiente buscado.

**Salida**: Un enunciado que declare $SÍ$ en caso de que el grafo contenga un conjunto independiente de tamaño $k$, y $NO$ en caso contrario.

---

#### **Pre-proceso**
Para comenzar la reducción del problema, primeramente debemos traducir la entrada del problema $A$ en una entrada del problema $B$. Puesto que la entrada para el problema IS es un grafo con nodos y arcos, y la entrada para SSP es un número entero positivo, debemos encontrar alguna forma de relacionar ambas y que a la vez puedan utilizarse para resolver el problema IS.

Para encontrar esta relación, primero debemos de darnos cuenta que encontrar el conjunto independiente con $k$ nodos, no es más que encontrar $k$ nodos que no sean conectados por el mismo arco. Ya tenemos como primera regla considerar tan solo una vez un mismo arco y mientras se realiza lo anterior, como segunda regla sería considerar una cantidad de exactamente de $k$ nodos, entonces podemos hacer la siguiente transformación de nodos a enteros para cumplir simultáneamente ambas reglas:

*   Cada nodo y cada arco será un número de $|E|+1$ dígitos.
*   Cada dígitos podrá ser un 1 o un 0.
*   El primer dígito será para identificar si es un nodo o un arco (se marcará con 1 o 0 respectivamente).
*   Desde el segundo dígito en adelante podrá ser un 1 si considera el arco $E_{dígito}$ o bien 0 si no lo considera. Por ejemplo cada arco $E_i$ se considera a sí mismo, entonces el i-ésimo dígito más 1 será el único con un 1 y cada nodo considera los arcos que lo conectan.
*   Por último, el valor objetivo $T$ será igual a un número de $|E|+1$ dígitos, donde el primer dígito ya no podrá ser solo un 0 o un 1, sino que representará el valor original de $k$ nodos considerados, y el resto de $E$ dígitos será 1.

### **Problema de la mochila**

Dados los pesos y valores de $i$ artículos, coloque estos artículos en una mochila de capacidad W para obtener el valor total máximo en la mochila.

---
**Entrada**: Conjunto de valores $value = \{v_1, v_2, v_3, ..., v_i\}$ y pesos $weight = \{w_1, w_2, w_3, ..., w_i\}$ asociados a una cantidad $i$ de artículos de forma respectiva, además de una capacidad máxima de peso soportado $W$.

**Salida**: el subconjunto de valor máximo de $value$ tal que la suma de los pesos de este subconjunto sea menor o igual a $W$.

---

### **Teorema**
>*El problema de la mochila es **NP-Hard**.*

Como se explicó anteriormente, para demostrar que el problema de la mochila es $NP-Hard$, debemos lograr reducir un problema $A \in NP-Hard$ al problema de la mochila. El problema $A$ escogido es la suma de subconjuntos (SSP), y asumiremos que este es $NP-Hard$.

### **Lema**
*El problema de la suma de subconjuntos se reduce al problema de la mochila.*

### **Prueba del lema**

El problema de la suma de subconjuntos ya fue descrito anteriormente, entonces a continuación describiremos que pre y post procesos realizarles al problema de la mochila. Como observación, se escogerá la variante en la que todas las entradas son positivas y la salida para este es por decisión, es decir, $SÍ$ si existe subconjunto o $NO$ si no existe.

#### **Pre-proceso**
Los pasos a realizarse para lograr la reducción es hacer un pre-proceso a la entrada de valores del SSP para convertirlos en una entrada para el problema de la mochila, es decir, la cantidad de valores $Wi$ de SSP serán la cantidad de artículos para el problema de la mochila, y como valor y peso de cada artículo serán el valor del entero en el conjunto SSP, $value_i = W_i$ y $weight_i = W_i$. También SSP recibe como entrada un valor objetivo $T$, lo que se traduce en el problema de la mochila como la capacidad máxima $WEIGHT = T$.

#### **Subrutina**
Se resuelve el problema de la mochila con la entrada anteriormente descrita, este buscará maximizar la gnanacia acercándose al valor objetivo, mientras el peso no sobrepase de igual manera el valor objetivo.

#### **Post-proceso**
Como salida del algoritmo podemos tener dos casos con respecto al conjunto que suma el valor máximo de beneficio:

*   Puede ser exactamente el valor objetivo del SSP, por lo que este algoritmo de decisión retornaría $SÍ$
*   Puede ser menor al valor objetivo del SSP o bien ser vacío, por lo que este algoritmo de decisión retornaría $NO$

Esto sucede porque el problema de la mochila prueba cada combinación de artículos para llegar al valor objetivo con los pesos y valores a la vez. **Solo si existe** alguna combinación que produzca este valor objetivo se retornará tal conjunto, **pero si no existe** retonará cualquier otro conjunto de valores que maximizan el beneficio, pero no llegan exactamente a la capacidad máxima o ningún valor en caso de que todos sean mayor a la capacidad de la mochila.

##Problema de la suma de subconjuntos 

**Algorimo de aproximacion de tiempo polinomial:** Suponiendo que los valores de entrada son positivos podemos describir el algoritmo. Cumple la funcion de encontrar un subconjunto de $A$ con un valor que representa la suma maxima $M$ y al menos una cantidad $R$ de sumas optimas ($R = 0,1$ relacion de aproximacion). 

Algoritmo que tiene una proporcion de 0,5, su funcionamiento se basa en ordenar las entradas por valores descendentes, para despues colocar la siguiente entrada mas grande en el subconjunto, siempre y cuando este pueda ingresarse dentro.

Este algoritmo al momento de finalizar, considera dos casos, siendo el primero un subconjunto optimo o un subconjunto en el que una de sus entradas no fue posible de ingresar. 

**Exclusion e Inclusion:** Es un algoritmo que puede ser implementado mediante la operacion de busqueda en profundidad de un arbol binario. Donde cada valor de entrada sera un nivel de nuestro arbol, la rama izquierda excluye el numero del conjunto mientras que la rama derecha incluye dicho valor. 

Este algoritmo funciona mediante los siguientes pasos:

1. Ordenar los numeros de entradas de forma descendentes.
2. Se pregunta si los numeros que fueron incluidos en un nodo dado del arbol $A$, exceden la suma del mejor subconjunto en ese momento $B$, si $A > B$ se corta el nodo $A$.
3. Se pregunta si los numeros incluidos en un nodo dado $A$, mas la suma de todos los numeros faltantes $A + C$, son menores a la suma del mejor subconjunto en ese momento $B$, si $(A + C) < B$ se corta $A$.   


##Problema de la mochila

**Algoritmo voraces:** Consiste en introducir los valores de la mochila de forma descendente de utilidad de los diversos objetos. Como primer paso se añadiran valores numericos hasta que por reestricciones de capacidad no sea posible el ingreso de mas valores, dando asi el caso en que debamos de ingresar una porcion del valor numerico que haga entrar el siguiente objeto.

Para conseguir su solucion "optima" al igual que en su solucion base ingresamos los valores de forma descendente en relacion a su utilidad y pondereacion/valor, para despues ingresar a la mochila valores numericos mientras que no exceda su capacidad, en caso de llegar a su limite donde un valor numerico no cabe en su totalidad se ingresa una porcion de si mismo que si lo haga dando asi un resultado optimo.
